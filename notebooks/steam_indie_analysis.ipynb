```python
# CELL 6 - Filter to Indie Games Only

# Keep only games that have 'Indie' in their genres list

indie_df=df[df['genres'].apply(lambda x: 'Indie' in x)].copy()

print(f"Original dataset: {len(df):,} games")
print(f"Indie games only: {len(indie_df):,} games")
print(f"Dropped: {len(df) - len(indie_df):,} non-indie games")
```

```python
# ============================================================
# CELL 7 — Parse estimated_owners into midpoint number
# ============================================================

def parse_owners(owners_str):
    # Split "50000 - 100000" into ["50000 ", " 100000"]
    parts = owners_str.split('-')
    
    # Convert each part to integer after stripping whitespace
    low = int(parts[0].strip())
    high = int(parts[1].strip())
    
    # Return the midpoint
    return (low + high) / 2

# Apply to every row
indie_df['owners_midpoint'] = indie_df['estimated_owners'].apply(parse_owners)

# Verify
print(indie_df[['estimated_owners', 'owners_midpoint']].head(10))
```

```python
# ============================================================
# CELL 8 — Calculate positive_rate
# ============================================================

# positive / (positive + negative)
# But if a game has 0 reviews, we'd get division by zero
# So we use a safe formula — if total is 0, rate is 0

total_reviews = indie_df['positive'] + indie_df['negative']

indie_df['positive_rate'] = indie_df['positive'] / total_reviews.replace(0, float('nan'))
indie_df['positive_rate'] = indie_df['positive_rate'].fillna(0)

# Verify
print(indie_df[['positive', 'negative', 'positive_rate']].head(10))
print(f"\nPositive rate stats:")
print(indie_df['positive_rate'].describe())
```

```python
# ============================================================
# CELL 9 — Estimate Revenue
# ============================================================

# estimated_revenue = owners_midpoint × price × 0.7
# 0.7 accounts for Steam's 30% cut

indie_df['estimated_revenue'] = (
    indie_df['owners_midpoint'] * 
    indie_df['price'] * 
    0.7
)

# Quick look at the distribution
print("Revenue stats (USD):")
print(indie_df['estimated_revenue'].describe())

print(f"\nGames with $0 estimated revenue: {(indie_df['estimated_revenue'] == 0).sum():,}")
print(f"Games with >$50k revenue: {(indie_df['estimated_revenue'] >= 50000).sum():,}")
print(f"Games with >$100k revenue: {(indie_df['estimated_revenue'] >= 100000).sum():,}")
```

```python
# ============================================================
# CELL 10 — Define Success Label
# ============================================================

# Success = estimated_revenue >= $50,000 AND positive_rate >= 0.70
indie_df['success'] = (
    (indie_df['estimated_revenue'] >= 50000) & 
    (indie_df['positive_rate'] >= 0.70)
).astype(int)

# Check class balance
success_counts = indie_df['success'].value_counts()
success_pct = indie_df['success'].value_counts(normalize=True) * 100

print("=== CLASS BALANCE ===")
print(f"Unsuccessful (0): {success_counts[0]:,} games ({success_pct[0]:.1f}%)")
print(f"Successful   (1): {success_counts[1]:,} games ({success_pct[1]:.1f}%)")
print(f"\nImbalance ratio: {success_counts[0]/success_counts[1]:.1f}:1")
```

```python
# ============================================================
# VISUAL 3 — Success Rate by Genre (Improved)
# ============================================================

meaningful_genres = [
    'Casual', 'Action', 'Adventure', 'Simulation', 
    'Strategy', 'RPG', 'Early Access', 'Free To Play',
    'Sports', 'Racing', 'Massively Multiplayer'
]

genre_stats = []
for genre in meaningful_genres:
    mask = indie_df['genres'].apply(lambda x: genre in x)
    total = mask.sum()
    successful = indie_df[mask]['success'].sum()
    rate = successful / total * 100
    genre_stats.append({
        'genre': genre,
        'total': total,
        'successful': successful,
        'success_rate': rate
    })

genre_df = pd.DataFrame(genre_stats).sort_values('success_rate', ascending=True)

# Color by performance tier
colors = []
for rate in genre_df['success_rate']:
    if rate >= 19:
        colors.append('#00b894')  # green — top tier
    elif rate <= 12:
        colors.append('#ff7675')  # red — low tier
    else:
        colors.append('#74b9ff')  # blue — mid tier

fig = go.Figure()

fig.add_trace(go.Bar(
    x=genre_df['success_rate'],
    y=genre_df['genre'],
    orientation='h',
    marker_color=colors,
    marker_line_width=0,
    opacity=0.90,
    text=[f"{r:.1f}%" for r in genre_df['success_rate']],
    textposition='outside',
    textfont=dict(color='#2d3436', size=11),
    hovertemplate='<b>%{y}</b><br>Success Rate: %{x:.1f}%<br>Total Games: %{customdata:,}<extra></extra>',
    customdata=genre_df['total']
))

# Add average line
avg = genre_df['success_rate'].mean()
fig.add_vline(
    x=avg,
    line_dash='dash',
    line_color='#636e72',
    line_width=1.5,
    annotation_text=f'Average: {avg:.1f}%',
    annotation_font=dict(color='#636e72', size=11),
    annotation_position='top'
)

fig.update_layout(
    title=dict(
        text='Success Rate by Genre — Indie Games',
        font=dict(size=17, color='#2d3436'),
        x=0.5,
        xanchor='center'
    ),
    xaxis=dict(
        title='Success Rate (%)',
        showgrid=True,
        gridcolor='#dfe6e9',
        gridwidth=1,
        color='#636e72',
        range=[0, max(genre_df['success_rate']) * 1.25]
    ),
    yaxis=dict(
        showgrid=False,
        color='#2d3436',
        tickfont=dict(size=12)
    ),
    plot_bgcolor='white',
    paper_bgcolor='white',
    height=550,
    width=750,
    margin=dict(l=20, r=80, t=60, b=50),
    annotations=[
        dict(
            x=0.98, y=0.02,
            xref='paper', yref='paper',
            text='🟢 Top tier  🔵 Mid tier  🔴 Low tier',
            showarrow=False,
            font=dict(size=10, color='#636e72'),
            align='right'
        )
    ]
)

fig.write_image('../images/visual3_success_rate_by_genre.png', scale=2)
fig.show()
```
