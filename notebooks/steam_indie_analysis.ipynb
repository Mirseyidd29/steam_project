```python
# Imports 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
plt.style.use('seaborn-v0_8')

#Display settings
pd.set_option('display.max_columns', 50)
pd.set_option('display.max_rows', 100)


print('All imports successful')
```

```python
# Loading data from HuggingFace


df=pd.read_parquet("hf://datasets/FronkonGames/steam-games-dataset/data/train-00000-of-00001.parquet")

print(f"Data loaded successfully")
print(f"Shape : {df.shape[0]:,} rows x {df.shape[1]} columns")
```

```python
# First look

#first 3 rows
print("===FIRST 3 ROWS ===")
display(df.head(3))


#columns and data types
print("\n=== COLUMNS & DTYPES ===")
print(df.dtypes)



# How much data is missing?
print("\n=== MISSING VALUES (top 15) ===")
missing = df.isnull().sum().sort_values(ascending=False)
print(missing[missing > 0].head(15))
```

```python
# What does estimated_owners look like?
print("=== ESTIMATED OWNERS (unique values sample) ===")
print(df['estimated_owners'].value_counts())
```

```python
# ============================================================
# CELL 5 — Inspect Genres & Tags
# ============================================================

print("=== GENRES (first 5) ===")
print(df['genres'].head())

print("\n=== TAGS (first 3) ===")
print(df['tags'].head(3))

print("\n=== RELEASE DATE (first 5) ===")
print(df['release_date'].head())

print("\n=== PRICE DISTRIBUTION ===")
print(df['price'].describe())
```

```python
indie_count = df['genres'].apply(lambda x: 'Indie' in x).sum()
total = len(df)

print(f"Total games: {total:,}")
print(f"Indie games: {indie_count:,}")
print(f"Percentage: {indie_count/total*100:.1f}%")
```
