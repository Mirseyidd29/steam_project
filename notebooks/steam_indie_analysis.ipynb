```python
indie_count = df['genres'].apply(lambda x: 'Indie' in x).sum()
total = len(df)

print(f"Total games: {total:,}")
print(f"Indie games: {indie_count:,}")
print(f"Percentage: {indie_count/total*100:.1f}%")
```

```python
# CELL 6 - Filter to Indie Games Only

# Keep only games that have 'Indie' in their genres list

indie_df=df[df['genres'].apply(lambda x: 'Indie' in x)].copy()

print(f"Original dataset: {len(df):,} games")
print(f"Indie games only: {len(indie_df):,} games")
print(f"Dropped: {len(df) - len(indie_df):,} non-indie games")
```

```python
# ============================================================
# CELL 7 — Parse estimated_owners into midpoint number
# ============================================================

def parse_owners(owners_str):
    # Split "50000 - 100000" into ["50000 ", " 100000"]
    parts = owners_str.split('-')
    
    # Convert each part to integer after stripping whitespace
    low = int(parts[0].strip())
    high = int(parts[1].strip())
    
    # Return the midpoint
    return (low + high) / 2

# Apply to every row
indie_df['owners_midpoint'] = indie_df['estimated_owners'].apply(parse_owners)

# Verify
print(indie_df[['estimated_owners', 'owners_midpoint']].head(10))
```

```python
# ============================================================
# CELL 8 — Calculate positive_rate
# ============================================================

# positive / (positive + negative)
# But if a game has 0 reviews, we'd get division by zero
# So we use a safe formula — if total is 0, rate is 0

total_reviews = indie_df['positive'] + indie_df['negative']

indie_df['positive_rate'] = indie_df['positive'] / total_reviews.replace(0, float('nan'))
indie_df['positive_rate'] = indie_df['positive_rate'].fillna(0)

# Verify
print(indie_df[['positive', 'negative', 'positive_rate']].head(10))
print(f"\nPositive rate stats:")
print(indie_df['positive_rate'].describe())
```

```python
# ============================================================
# CELL 9 — Estimate Revenue
# ============================================================

# estimated_revenue = owners_midpoint × price × 0.7
# 0.7 accounts for Steam's 30% cut

indie_df['estimated_revenue'] = (
    indie_df['owners_midpoint'] * 
    indie_df['price'] * 
    0.7
)

# Quick look at the distribution
print("Revenue stats (USD):")
print(indie_df['estimated_revenue'].describe())

print(f"\nGames with $0 estimated revenue: {(indie_df['estimated_revenue'] == 0).sum():,}")
print(f"Games with >$50k revenue: {(indie_df['estimated_revenue'] >= 50000).sum():,}")
print(f"Games with >$100k revenue: {(indie_df['estimated_revenue'] >= 100000).sum():,}")
```

```python
# ============================================================
# CELL 10 — Define Success Label
# ============================================================

# Success = estimated_revenue >= $50,000 AND positive_rate >= 0.70
indie_df['success'] = (
    (indie_df['estimated_revenue'] >= 50000) & 
    (indie_df['positive_rate'] >= 0.70)
).astype(int)

# Check class balance
success_counts = indie_df['success'].value_counts()
success_pct = indie_df['success'].value_counts(normalize=True) * 100

print("=== CLASS BALANCE ===")
print(f"Unsuccessful (0): {success_counts[0]:,} games ({success_pct[0]:.1f}%)")
print(f"Successful   (1): {success_counts[1]:,} games ({success_pct[1]:.1f}%)")
print(f"\nImbalance ratio: {success_counts[0]/success_counts[1]:.1f}:1")
```
